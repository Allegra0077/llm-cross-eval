```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
path = sorted(Path("../results").glob("exp1_turn1_*.csv"))[-1] #pick newest exp1 csv
df = pd.read_csv(path)
path, df.head()

#plot
for pid, g in df.groupby("prompt_id"):
    plt.figure(figsize=(10,6))
    plt.title(f"Prompt {pid} - Token-level pA/pB ratios")
    plt.scatter(g["token_idx"], g["ratio_pA_over_pB"], s=12)
    plt.axhline(g["ratio_pA_over_pB"].mean(), linewidth=1)  # average line
    plt.xlabel("Token Index in Continuation")
    plt.ylabel("pA / pB")
    plt.yscale("log")
    plt.legend()
    plt.grid(True)
    plt.show()
```
