```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

results_dir = Path("../results")

# pick the newest exp2 csv
files = sorted(results_dir.glob("exp2_canonical_*.csv")) 
path = files[-1]

df = pd.read_csv(path)
path, df.shape
```

```python
plt.figure(figsize=(10,6))
plt.title("Canonical text â€” Token-level pA/pB ratios")

plt.scatter(
    df["token_idx"],
    df["ratio_pA_over_pB"],
    s=12
)

plt.xlabel("Token index in text")
plt.ylabel("pA / pB")
plt.yscale("log")
plt.legend()
plt.grid(True)
plt.show()
```
